```python
X_train = train_df.drop(columns=["default.payment.next.month"])
X_test = test_df.drop(columns=["default.payment.next.month"])

y_train = train_df["default.payment.next.month"]
y_test = test_df["default.payment.next.month"]

numeric_features = [
    "LIMIT_BAL", 
    "AGE",
    "BILL_AMT1",
    "BILL_AMT2",
    "BILL_AMT3",
    "BILL_AMT4",
    "BILL_AMT5",
    "BILL_AMT6",
    "PAY_AMT1",
    "PAY_AMT2",
    "PAY_AMT3",
    "PAY_AMT4",
    "PAY_AMT5",
    "PAY_AMT6"
]
categorical_features = ["EDUCATION","MARRIAGE", "PAY_0", "PAY_2", "PAY_3", "PAY_4", "PAY_5", "PAY_6"]
binary_features = ["SEX"]
target = "default.payment.next.month"

numeric = make_pipeline(
    StandardScaler()
)

categorical = make_pipeline(
    OneHotEncoder(handle_unknown="ignore", sparse=False)
)

binary = make_pipeline(
    OneHotEncoder(drop="if_binary", dtype=int)
)

preprocessor = make_column_transformer(
    (numeric, numeric_features),
    (categorical, categorical_features),
    (binary, binary_features)
)
```

## 6. Baseline model <a name="6"></a>
<hr>

rubric={points:2}

**Your tasks:**
1. Try `scikit-learn`'s baseline model and report results.

```python
def mean_std_cross_val_scores(model, X_train, y_train, **kwargs):
    scores = cross_validate(model, X_train, y_train, **kwargs)

    mean_scores = pd.DataFrame(scores).mean()
    std_scores = pd.DataFrame(scores).std()
    out_col = []

    for i in range(len(mean_scores)):
        out_col.append((f"%0.3f (+/- %0.3f)" % (mean_scores[i], std_scores[i])))

    return pd.Series(data=out_col, index=mean_scores.index)

baseline_model = make_pipeline(
    preprocessor,
    DummyClassifier(strategy="prior")
)
results_dict = {}
results_dict["baseline"] = mean_std_cross_val_scores(baseline_model, X_train, y_train, cv=5, return_train_score=True)
pd.DataFrame(results_dict)
```

```python
# 2. carry out hyper param tuning
from scipy.stats import loguniform

param_dist = {"logisticregression__C": loguniform(1e-3, 1e3)}
pipe_logit = make_pipeline(preprocessor, LogisticRegression(max_iter=2000))
random_search = RandomizedSearchCV(
    pipe_logit, param_distributions=param_dist, n_jobs=-1, cv=5, random_state=123
)
random_search.fit(X_train, y_train)
```

```python
# 3. report the mean validation scores along with standard deviation
pd.DataFrame(random_search.cv_results_)[
    [
        "mean_test_score",
        "std_test_score",
        "param_logisticregression__C",
        "mean_fit_time",
        "rank_test_score",
    ]
].set_index("rank_test_score").sort_index().T
```

4. I used randomized search rather than gridSearchCV because it's faster and easier to locate the useful hyperparams than gridSearchCV. After the randomized search, we show that C=766 is the best with highest mean_test_score. However, I think C=12.85 should be instead adopted because it yields the second highest mean_test_score (and their scores do not differ by much) and it's way less complex than C=766. ## 11. Interpretation and feature importances <a name="1"></a>
<hr>
rubric={points:15}

**Your tasks:**

1. Use the methods we saw in class (e.g., `eli5`, `shap`) (or any other methods of your choice) to explain feature importances of one of the best performing models. Summarize your observations. 

```python
# adopted from lecture 12
import eli5
from lightgbm.sklearn import LGBMClassifier
best_lgbm_pipe = make_pipeline(preprocessor, LGBMClassifier(random_state = 123, num_leaves = 26))
best_lgbm_pipe.fit(X_train, y_train)
```

```python
categorical = best_lgbm_pipe.named_steps['columntransformer'].transformers_[1][1]['onehotencoder'].get_feature_names_out().tolist()
binary = best_lgbm_pipe.named_steps['columntransformer'].transformers_[2][1]['onehotencoder'].get_feature_names_out().tolist()
feature_names = numeric_features + categorical + binary
eli5.explain_weights(
    best_lgbm_pipe.named_steps["lgbmclassifier"], feature_names = feature_names
)
```

```python
preprocessor2 = make_column_transformer(
    (StandardScaler(), numeric_features),
    (OneHotEncoder(handle_unknown="ignore", sparse=False), categorical_features),
    (OneHotEncoder(drop="if_binary", dtype=int), binary_features)
)
preprocessor2.fit(X_train, y_train)
X_train_enc = pd.DataFrame(
    data=np.array(preprocessor2.transform(X_train)),
    columns=feature_names,
    index=X_train.index,
)
print(X_train_enc.shape)
X_test_enc = pd.DataFrame(
    data=np.array(preprocessor2.transform(X_test)),
    columns=feature_names,
    index=X_test.index,
)

print(X_test_enc.shape)
import shap

lgbm_explainer = shap.TreeExplainer(best_lgbm_pipe.named_steps["lgbmclassifier"])
train_lgbm_shap_values = lgbm_explainer.shap_values(X_train_enc)
values = np.abs(train_lgbm_shap_values[1]).mean(0)
pd.DataFrame(data=values, index=feature_names, columns=["SHAP"]).sort_values(
    by="SHAP", ascending=False
)[:10]
```

As we can see from the above using both eli5 and shap, we find that "PAY_0_2" is the most important feature both in terms of eli5 and shap and its importance is way higher than any other features. Moreover, some features are importants that they appear in the both the top 10 in eli5 and shap, namely PAY_0_2, LIMIT_BAL, PAY_2_2, PAY_AMT2, BILL_AMT1, PAY_AMT1, PAY_AMT3, PAY_3_2 and PAY_AMT4.
